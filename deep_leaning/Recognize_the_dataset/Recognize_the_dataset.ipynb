```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python


# 从torchvision.datasets中加载FashionMNIST数据集
train_data = datasets.FashionMNIST(
    root='../data',  # 数据存储路径
    train=True,   # 使用训练集
    download=True,  # 如果数据不存在则下载
    transform=ToTensor()  # 将图像转换为张量
)

# 加载测试数据集
test_data = datasets.FashionMNIST(
    root='../data',  # 指定数据存储路径
    train=False,  # 使用测试集
    download=True,  # 如果数据不存在则下载
    transform=ToTensor()  # 将图像转换为张量
)

# FashionMNIST数据集的类别标签
labels_map = {
    0: "T-shirt",    # T恤
    1: "Trouser",    # 裤子
    2: "Pullover",   # 套头衫
    3: "Dress",      # 连衣裙
    4: "Coat",       # 外套
    5: "Sandal",     # 凉鞋
    6: "Shirt",      # 衬衫
    7: "Sneaker",    # 运动鞋
    8: "Bag",        # 包
    9: "Ankle boot"  # 短靴
}

# 显示数据集中的一些样本图片
figure = plt.figure(figsize=(8, 8))  # 创建8x8大小的图形
cols, rows = 3, 3  # 设置显示3x3的网格
for i in range(1, cols * rows + 1):  # 遍历每个网格位置
    sample_idx = torch.randint(len(train_data), size=(1,)).item()  # 随机选择一个样本索引
    img, label = train_data[sample_idx]  # 获取图像和标签
    figure.add_subplot(rows, cols, i)  # 添加子图
    plt.title(labels_map[label])  # 设置子图标题为类别名称
    plt.axis("off")  # 关闭坐标轴显示
    plt.imshow(img.squeeze(), cmap="gray")  # 显示灰度图像
plt.show()  # 显示整个图形

# 打印数据集基本信息
print(f"训练集大小: {len(train_data)}")  # 打印训练集样本数量
print(f"测试集大小: {len(test_data)}")  # 打印测试集样本数量
print(f"图像大小: {train_data[0][0].shape}")  # 打印单个图像的形状

# 查看训练数据的形状
print(f"训练数据的形状: {train_data.data.shape}")  # 打印整个训练集的形状
print(f"训练数据标签的形状: {train_data.targets.shape}")  # 打印训练集标签的形状
# 获取第一张图片并打印其形状
first_image, _ = train_data[0]
print(f"第一张图片的形状: {first_image.shape}")
print("#"*100)
# 打印第一张图片的像素值，这是一个28x28的张量，每个值表示对应像素点的灰度值（0-1之间）
print(train_data[0][0])

```
